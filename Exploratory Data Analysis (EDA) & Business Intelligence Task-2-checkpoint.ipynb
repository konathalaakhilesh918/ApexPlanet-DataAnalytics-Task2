```python
import pandas as pd
df = pd.read_csv("C:/Users/hp/.ipynb_checkpoints/cleaned_dataset-checkpoint.csv")
df.head()
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
# STEP 1: Descriptive Statistics (EDA Basics)
df.info()
df.describe()
df.isnull().sum()
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Histogram (Sales Distribution)
plt.hist(df['Total_Sales'], bins=30)
plt.title("Sales Distribution")
plt.show()
```

```python
#2. Sales by Category
df.groupby('Category')['Total_Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
#3. Sales by City
df.groupby('City')['Total_Sales'].sum().sort_values(ascending=False).head().plot(kind='bar')
plt.title("Top Cities by Sales")
plt.show()
```

```python
#STEP 2: Business Questions + SQL
print("Q1: Top 5 Products by Sales")
df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False).head()
```

```python
print("Q2: Monthly Sales Trend")
df.groupby('Order_Month')['Total_Sales'].sum()
```

```python
print("Q3: Which Category Gives Highest Sales?")
df.groupby('Category')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
print("Q4: Which Gender Spends More?")
df.groupby('Gender')['Total_Sales'].sum()
```

```python
print("Q5: High Value Orders Count")
df['High_Value_Order'].value_counts()
```

```python
#STEP 3: Multivariate Analysis
#Scatter Plot (Sales vs Quantity)
plt.scatter(df['Quantity'], df['Total_Sales'])
plt.xlabel("Quantity")
plt.ylabel("Sales")
plt.title("Quantity vs Sales")
plt.show()
```

```python
#Correlation Heatmap
numeric_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
